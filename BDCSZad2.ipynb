```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.model_selection import train_test_split
```

```python
path = "sample_data/CharlesBookClub.csv"
df = pd.read_csv(path)

print(df.shape)
print(df.head())
```

R - Recency (Months Since last purchase)
F - Frequency - Total number of purchase


the segmentation process aims to divide customers into homogeneous groups. The homogeneous criterion we need for segmentation is the probability to purchase the offering. In this project,  use 3 variables R, F, M as indicators of this propensity

```python
#Split 60/40 na train/test - to be used later
train_df, test_df = train_test_split(df, test_size=0.4, random_state=123, shuffle=True)
```

# Segmentację od teraz robię na danych treningowych - na poźniej

```python
# czy klient kupił Florence - odsetek jedynek = bazowy response rate
# do użycia w określeniu rr
m = (train_df["Florence"] == 1).mean()
m
```

# podanie wszystkich segmentów R/F/M z obliczeniem odsetek zakupionych Florence - UWAGA dla Treningowego

```python
#podział na segmenty i sumowanie jaka kategoria ma jaką średnią (ile kupiła książek Florence)
rr = train_df.groupby(["Rcode","Fcode","Mcode"])["Florence"].mean().reset_index(name="rr")
rr
```

```python
# wykres słupkowy, żeby złapać ooziom

rr_plot = rr.copy()  # kopia
rr_plot["RFM"] = rr_plot["Rcode"].astype(str) + rr_plot["Fcode"].astype(str) + rr_plot["Mcode"].astype(str)  # etykieta segmentu
rr_mean = rr_plot["rr"].mean()  # mean(rr)

fig = px.bar(
    rr_plot.sort_values("rr", ascending=False),  # sortuj malejąco
    x="RFM", y="rr",  # osie
    title="RFM względem mean()",  # tytuł
    template="plotly_dark"  # dark
)

fig.add_hline(y=rr_mean, annotation_text=f"mean(rr)={rr_mean:.4f}", annotation_position="top left")  # linia średniej
fig.update_layout(xaxis_title="Segment RFM", yaxis_title="rr", xaxis_tickangle=-90)  # czytelność
fig.show()


```

dyskretyzacja / kategoryzowanie : z ciągłych (lub liczbowych) R/F/M robie małe, skończone klasy, żeby potem grupować segmenty typu RFM

```python
# jawne przyporządkowanie -> Rcode/Fcode/Mcode

train_df["Rcode"] = np.select(
    [train_df["R"].between(0, 2),
     train_df["R"].between(3, 6),
     train_df["R"].between(7, 12),
     train_df["R"] >= 13],
    [1, 2, 3, 4],
    default=0
).astype(int)

train_df["Fcode"] = np.select(
    [train_df["F"] == 1,
     train_df["F"] == 2,
     train_df["F"] >= 3],
    [1, 2, 3],
    default=0
).astype(int)

train_df["Mcode"] = np.select(
    [train_df["M"].between(0, 25),
     train_df["M"].between(26, 50),
     train_df["M"].between(51, 100),
     train_df["M"].between(101, 200),
     train_df["M"] >= 201],
    [1, 2, 3, 4, 5],
    default=0
).astype(int)

```

#kombinacje dla segmentu #1

Segment 1 - kombinacje RFM, dla których rr > 2 × (overall response rate) (liczone na zbiorze treningowym).
- wyszła pusta tabela - nie ma kombinacji

```python
#wszystkie kombinacje R/F/M dla segmentu 1 (wg rr > 2*overall_rr)
overall_rr = train_df["Florence"].mean()  # baseline (train)

rr = (train_df.groupby(["Rcode", "Fcode", "Mcode"])["Florence"]
      .mean()
      .reset_index(name="rr"))

seg1 = (rr.loc[rr["rr"] > 2 * overall_rr, ["Rcode", "Fcode", "Mcode", "rr"]]
        .assign(RFM=lambda d: d["Rcode"].astype(str) + d["Fcode"].astype(str) + d["Mcode"].astype(str))
        .sort_values("rr", ascending=False)
        .reset_index(drop=True))

seg1
```

#kombinacje dla segmentu #2

Segment 2 to kombinacje RFM -  overall response rate < rr ≤ 2 × overall response rate (na train).

```python
# wszystkie kombinacje R/F/M dla segmentu 2 (overall_rr < rr <= 2*overall_rr)
overall_rr = train_df["Florence"].mean()  # baseline (train)

rr = (train_df.groupby(["Rcode", "Fcode", "Mcode"])["Florence"]
      .mean()
      .reset_index(name="rr"))

seg2 = (rr.loc[(rr["rr"] > overall_rr) & (rr["rr"] <= 2 * overall_rr), ["Rcode", "Fcode", "Mcode", "rr"]]
        .assign(RFM=lambda d: d["Rcode"].astype(str) + d["Fcode"].astype(str) + d["Mcode"].astype(str))
        .sort_values("rr", ascending=False)
        .reset_index(drop=True))

seg2
```

Segment 3 to reszta, czyli rr ≤ overall response rate (pozostałe kombinacje RFM).

```python
# foreword: wszystkie kombinacje R/F/M dla segmentu 3 (rr <= overall_rr)
overall_rr = train_df["Florence"].mean()  # baseline (train)

rr = (train_df.groupby(["Rcode", "Fcode", "Mcode"])["Florence"]
      .mean()
      .reset_index(name="rr"))

seg3 = (rr.loc[rr["rr"] <= overall_rr, ["Rcode", "Fcode", "Mcode", "rr"]]
        .assign(RFM=lambda d: d["Rcode"].astype(str) + d["Fcode"].astype(str) + d["Mcode"].astype(str))
        .sort_values("rr", ascending=False)
        .reset_index(drop=True))

seg3
```

Nr segmentu (1-3), który jest najbardziej obiecujący (zapewnia największy zysk) z punktu
widzenia sprzedaży książki The Art of History of Florence.

```python
#segment 1-3  wybór najbardziej obiecującego (najwyższy mean(rr))
overall_rr = train_df["Florence"].mean()  # bazowy response rate na train

rr = (train_df.groupby(["Rcode", "Fcode", "Mcode"])["Florence"]
      .mean()
      .reset_index(name="rr"))

rr["segment"] = np.select(  # segmentacja 1-3
    [rr["rr"] > 2 * overall_rr, rr["rr"] > overall_rr],
    [1, 2],
    default=3
).astype(int)

seg_stats = (rr.groupby("segment")["rr"]
             .mean()
             .reset_index(name="mean_rr")
             .sort_values("mean_rr", ascending=False)
             .reset_index(drop=True))

best_segment = int(seg_stats.loc[0, "segment"])  # nr segmentu 1-3
print("Najlepszy segment: ",best_segment)
print("-----------------------")
print(seg_stats)

```

```python
# lift chart na test_df; score = rr segmentu (Rcode/Fcode/Mcode); decyle 10%

overall_rr = train_df["Florence"].mean()  # bazowy rr (train) - do liftu

rr = (train_df.groupby(["Rcode", "Fcode", "Mcode"])["Florence"]
      .mean()
      .reset_index(name="rr"))  # score per RFM (train)

test_scored = (test_df.merge(rr, on=["Rcode", "Fcode", "Mcode"], how="left")
               .assign(score=lambda d: d["rr"].fillna(overall_rr)))  # fallback gdy brak segmentu

test_scored = test_scored.sort_values("score", ascending=False).reset_index(drop=True)  # ranking
test_scored["decile"] = (np.floor(np.arange(len(test_scored)) / (len(test_scored) / 10)) + 1).astype(int)  # 1..10
test_scored["decile"] = test_scored["decile"].clip(1, 10)  # safety

lift_tbl = (test_scored.groupby("decile")
            .agg(
                n_customers=("Florence", "size"),            # liczba klientów w decylu
                n_purchases=("Florence", "sum"),             # liczba zakupów (1)
                response_rate=("Florence", "mean"),          # rr w decylu
            )
            .reset_index()
            .assign(
                baseline_rr=overall_rr,                      # baseline (train)
                lift=lambda d: d["response_rate"] / d["baseline_rr"]  # lift = rr_decyl / baseline
            ))

fig = px.bar(
    lift_tbl,
    x="decile",
    y="lift",
    hover_data=["n_customers", "n_purchases", "response_rate", "baseline_rr"],
    title="Lift Chart (TEST) — score=rr(Rcode,Fcode,Mcode) from TRAIN; baseline=overall_rr(TRAIN)",
    template="plotly_dark"
)
fig.update_layout(
    xaxis_title="Decile (1 = najwyższy score, 10 = najniższy)",
    yaxis_title="Lift = response_rate(decile) / baseline_rr(train)"
)
fig.show()

lift_tbl
```

Reggresja Logistyczna - Wariant A

```python
# Logistic Regression (wariant A): wszystkie predyktory, scaler fit na train, proba na test
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# target
y_train = train_df["Florence"]  # y train
y_test  = test_df["Florence"]   # y test

# X: wszystko poza target + ID/duplikaty targetu
drop_cols = [c for c in ["Florence", "Seq#", "ID#", "Yes_Florence", "No_Florence"] if c in train_df.columns]
X_train = train_df.drop(columns=drop_cols)  #  train
X_test  = test_df.drop(columns=drop_cols)   #  test

# scaler: fit  na train
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std  = scaler.transform(X_test)


lr_all = LogisticRegression(max_iter=4000, solver="lbfgs")
lr_all.fit(X_train_std, y_train)


proba_test_all = lr_all.predict_proba(X_test_std)[:, 1]  # score
#wyświetl pierwsze 10 elementów wektora prawdopodobieństw
proba_test_all[:10]
```

Regresja logistyczna wariant B

```python
# Logistic Regression (wariant B): tylko R,F,M; scaler fit na train; proba na test
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# target
y_train = train_df["Florence"]  # y train
y_test  = test_df["Florence"]   # y test

# X tylko R,F,M
X_train_rfm = train_df[["R", "F", "M"]]  # cechy train
X_test_rfm  = test_df[["R", "F", "M"]]   # cechy test

# scaler: fit tylko na train
scaler_rfm = StandardScaler()  # standaryzacja
X_train_rfm_std = scaler_rfm.fit_transform(X_train_rfm)  # fit+transform train
X_test_rfm_std  = scaler_rfm.transform(X_test_rfm)       # transform test

# model
lr_rfm = LogisticRegression(max_iter=4000, solver="lbfgs")  # logreg
lr_rfm.fit(X_train_rfm_std, y_train)  # trening

# proba na test: P(Florence=1)
proba_test_rfm = lr_rfm.predict_proba(X_test_rfm_std)[:, 1]  # score

proba_test_rfm[:10]
```

```python
# confusion matrix + metryki dla wariantu A i B (threshold=0.5)
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

thr = 0.5  # próg klasyfikacji

# predykcje 0/1 z proba
y_pred_all = (proba_test_all >= thr).astype(int)  # wariant A
y_pred_rfm = (proba_test_rfm >= thr).astype(int)  # wariant B

# confusion matrix
cm_all = confusion_matrix(y_test, y_pred_all)  # [[TN,FP],[FN,TP]]
cm_rfm = confusion_matrix(y_test, y_pred_rfm)  # [[TN,FP],[FN,TP]]

# metryki (na TEST)
metrics_all = {
    "accuracy": accuracy_score(y_test, y_pred_all),
    "precision": precision_score(y_test, y_pred_all, zero_division=0),
    "recall": recall_score(y_test, y_pred_all, zero_division=0),
    "f1": f1_score(y_test, y_pred_all, zero_division=0),
    "roc_auc": roc_auc_score(y_test, proba_test_all),
}

metrics_rfm = {
    "accuracy": accuracy_score(y_test, y_pred_rfm),
    "precision": precision_score(y_test, y_pred_rfm, zero_division=0),
    "recall": recall_score(y_test, y_pred_rfm, zero_division=0),
    "f1": f1_score(y_test, y_pred_rfm, zero_division=0),
    "roc_auc": roc_auc_score(y_test, proba_test_rfm),
}

print(cm_all, metrics_all, cm_rfm, metrics_rfm)
```

```python
#confusion matrix w plotly (wariant A i B), dark, z opisami osi
import plotly.express as px

labels = ["0 (no)", "1 (yes)"]  # klasy Florence

cm_all_df = pd.DataFrame(cm_all, index=[f"Actual {l}" for l in labels], columns=[f"Pred {l}" for l in labels])  # A
cm_rfm_df = pd.DataFrame(cm_rfm, index=[f"Actual {l}" for l in labels], columns=[f"Pred {l}" for l in labels])  # B

fig_all = px.imshow(
    cm_all_df,
    text_auto=True,
    title="Confusion Matrix (TEST) — Logistic Regression (All predictors) @ threshold=0.5",
    template="plotly_dark"
)
fig_all.update_layout(xaxis_title="Predicted class", yaxis_title="Actual class")
fig_all.show()

fig_rfm = px.imshow(
    cm_rfm_df,
    text_auto=True,
    title="Confusion Matrix (TEST) — Logistic Regression (R,F,M only) @ threshold=0.5",
    template="plotly_dark"
)
fig_rfm.update_layout(xaxis_title="Predicted class", yaxis_title="Actual class")
fig_rfm.show()
```

```python
# lift chart (TEST) dla obu wariantów; baseline = rr_overall_test; decyle 10%
import numpy as np
import pandas as pd
import plotly.express as px

baseline_test = test_df["Florence"].mean()

def lift_table(y_true, proba, model_name):
    t = pd.DataFrame({"y": y_true.values, "proba": proba})
    t = t.sort_values("proba", ascending=False).reset_index(drop=True)
    t["decile"] = (np.floor(np.arange(len(t)) / (len(t) / 10)) + 1).astype(int)  # 1..10
    t["decile"] = t["decile"].clip(1, 10)
    out = (t.groupby("decile")
           .agg(customers=("y", "size"), purchases=("y", "sum"), rr_decile=("y", "mean"))
           .reset_index())
    out["baseline_rr_test"] = baseline_test
    out["lift"] = out["rr_decile"] / out["baseline_rr_test"]
    out["model"] = model_name
    return out

lift_all = lift_table(test_df["Florence"], proba_test_all, "LogReg — All predictors")  # wariant A
lift_rfm = lift_table(test_df["Florence"], proba_test_rfm, "LogReg — R,F,M only")      # wariant B

lift_long = pd.concat([lift_all, lift_rfm], ignore_index=True)  # long

fig = px.line(
    lift_long,
    x="decile",
    y="lift",
    color="model",
    markers=True,
    hover_data=["customers", "purchases", "rr_decile", "baseline_rr_test"],
    title="Lift Chart (TEST) — deciles by predicted probability; Lift = rr_decile / rr_overall_test",
    template="plotly_dark"
)
fig.update_layout(
    xaxis_title="Decile (1 = highest predicted probability, 10 = lowest)",
    yaxis_title="Lift = response_rate(decile) / overall_response_rate(TEST)"
)
fig.show()

lift_all, lift_rfm
```

```python
#cumulative gains chart (TEST) dla obu wariantów; sort po proba; porównanie do losowej
import numpy as np
import pandas as pd
import plotly.express as px

def gains_curve(y_true, proba, model_name):  # helper
    t = pd.DataFrame({"y": y_true.values, "proba": proba})  # tabela
    t = t.sort_values("proba", ascending=False).reset_index(drop=True)  # sort malejąco
    t["cum_pop"] = (np.arange(len(t)) + 1) / len(t)  # % populacji (0..1)
    total_pos = t["y"].sum()  # liczba pozytywów w test
    t["cum_tp"] = t["y"].cumsum()  # kumulacja TP
    t["cum_gain"] = (t["cum_tp"] / total_pos) if total_pos > 0 else 0.0  # % przechwyconych pozytywów
    out = t[["cum_pop", "cum_gain"]].copy()  # krzywa
    out["model"] = model_name  # tag
    return out

g_all = gains_curve(test_df["Florence"], proba_test_all, "LogReg — All predictors")  # A
g_rfm = gains_curve(test_df["Florence"], proba_test_rfm, "LogReg — R,F,M only")      # B

# linia losowa: y = x
g_rand = pd.DataFrame({"cum_pop": np.linspace(0, 1, 101), "cum_gain": np.linspace(0, 1, 101), "model": "Random (baseline)"} )

g_long = pd.concat([g_all, g_rfm, g_rand], ignore_index=True)  # long

fig = px.line(
    g_long,
    x="cum_pop",
    y="cum_gain",
    color="model",
    title="Cumulative Gains Chart (TEST) — captured positives vs population (sorted by predicted probability)",
    template="plotly_dark"
)
fig.update_layout(
    xaxis_title="Population fraction targeted (sorted by predicted probability)",
    yaxis_title="Cumulative gain = captured positives / total positives (TEST)"
)
fig.show()

g_all.head(), g_rfm.head()
```
